```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.model_selection import train_test_split
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('krisha.csv')
df =df[((df["District"]=="Бостандыкский")&(df["Size"]<=220))]
df=df[np.abs(df.Price-df.Price.mean()) <= (3.0*df.Price.std())]
df.shape
```

```python
y = df.select_dtypes(include=[np.number]).fillna(0)['Price']
x = df.select_dtypes(include=[np.number]).fillna(0).drop(columns=['Price'])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.5)
model = Ridge(alpha=.1)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
model.score(x_test, y_test)
```
