```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('train.csv')
```

Достанем из train_df признак SalePrice и обозначим его как целевой признак(зависимый).

Оставим только количественные признаки и разобьем данные на тренировочные и для проверки

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.model_selection import train_test_split

y = train_df.select_dtypes(include=[np.number]).fillna(0)['SalePrice']
X = train_df.select_dtypes(include=[np.number]).fillna(0).drop(columns=['SalePrice'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)
```

Используем линейную регрессию с л2-регуляризатором с параметром 0.1; То есть не сильно штрафуем за переобучение

```python
model = Ridge(alpha=.1)
model.fit(X_train, y_train)
res = model.predict(X_test)
mean_squared_error(res, y_test)
```

```python
model.score(X_test, y_test)
```

Построим heatmap корреляций между признаками. Как видим по диагонали конечно же стоят единицы и, возможно, между признаками есть зависимость. Но нас интересует зависимость между целевым признаком SalePrice и остальными

```python
corrs = num_train_df.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(corrs);
```

Оставим только те признаки, которые по модулю больше 0.5

```python
sp_name_corrs = corrs.index[abs(corrs['SalePrice'] > 0.5)]
plt.figure(figsize=(6, 6))
sp_corrs = train_df[sp_name_corrs].corr()
sns.heatmap(sp_corrs, annot=True);
```

Опять применяю регрессию только к тем признакам, от которых корелляция выше 0.5

```python
model = Ridge(alpha=.1)

y = train_df[sp_name_corrs]['SalePrice']
X = train_df[sp_name_corrs].drop(columns=['SalePrice'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)

model.fit(X_train, y_train)
pre_y = model.predict(X_test)
mean_squared_error(pre_y, y_test)
```

```python
model.score(X_test, y_test)
```

**Попробуем по-другому**

Уберем Id

```python
train_df = train_df.drop(columns='Id')
```

Уберем также те признаки, которые имеют корреляцию меньше 0.2

```python
corrs = train_df.corr()
corr_names = corrs.index[abs(corrs['SalePrice']) < 0.21]
train_df = train_df.drop(columns=corr_names)
```

Разобьем на X и y

```python
X = train_df.drop(columns='SalePrice')
y = train_df['SalePrice']
```

Посмотрим по каким признакам сколько не хватает данных

```python
usel_attr = X.isnull().sum() / X.shape[0] * 100
usel_attr = usel_attr[usel_attr > 0]
usel_attr.sort_values(inplace=True, ascending=False)
```

```python
usel_attr = usel_attr.to_frame()
```

```python
usel_attr.columns = ['percent']
usel_attr['Name'] = usel_attr.index
usel_attr.head()
```

```python
plt.figure(figsize=(13, 5))
plt.xticks(rotation=90)
sns.barplot(x='Name', y='percent', data=usel_attr);
```

Первые четыре признака почти пустые. Удалим их

```python
X.drop(columns=['PoolQC', 'MiscFeature', 'Alley', 'Fence'], inplace=True)
```

Заполним оставшиеся признаки

Все категориальные признаки заполним одинаковым значением 'None'

```python
for attr in X.select_dtypes(include='object'):
    X[attr].fillna('None', inplace=True)
```

А все числовые заполним медианой

```python
for attr in X.select_dtypes(include=['float64', 'int64']):
    X[attr].fillna(X[attr].median(), inplace=True)
```

Теперь с признаками можно произвести one-hot-encoding (ЭТО ПРИМЕНЯЕТСЯ КО ВСЕМ ПРИЗНАКАМ?)

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
X = ohe.fit_transform(X).toarray()
```

Разобьем на 80 на 20

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
model = Ridge(alpha=.1)
model.fit(X_train, y_train)
```

```python
pre_y = model.predict(X_test)
mean_squared_error(pre_y, y_test)
```

```python
model.score(X_test, y_test)
```

Запустим кросс-валидацию из 5 блоков и подберем параматр альфа

```python
from sklearn.model_selection import GridSearchCV

params = {'alpha': np.linspace(0.1, 1, 10)}
best = GridSearchCV(model, params, cv=5)
best.fit(X_train, y_train)
```
